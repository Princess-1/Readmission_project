```python
df.drop(['weight', 'payer_code', 'patient_nbr', 'medical_specialty', 'examide', 'citoglipton', 'glimepiride-pioglitazone', 'metformin-rosiglitazone', 
         'metformin-pioglitazone', 'acetohexamide', 'troglitazone' ], axis = 1, inplace = True)
df.head(10)
```

```python
df['readmitted'] = df['readmitted'].map(lambda x: 2 if x=='>30' else (1 if x=='<30' else 0))
df.readmitted = df.readmitted.astype('int64')
df.head()
```

```python
indexes_to_drop = []
for col in df:
    for index, row in enumerate(df[col]):
        #print(index, row)
        if row == '?':
            indexes_to_drop.append(index)
print(len(indexes_to_drop))
df.drop(index= indexes_to_drop, inplace = True)
df.shape
        
```
